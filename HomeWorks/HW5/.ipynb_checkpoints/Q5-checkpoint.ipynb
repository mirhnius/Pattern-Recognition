```python
def BFS(train_data, test_data, test_labels, itr, indices):
    dim = train_data.shape[1]
    CRR_BFS = np.zeros(itr)
    for j in range(itr):
        correctClassificationRate = np.zeros((dim))
        for i in range(dim):
            if i in indices:
                train = train_data
                labels = train_labels
                indx = indices.copy()
                indx.remove(i)
                gnb = MultinomialNB()
                t_labels = gnb.fit(train, labels).predict(test)
                where = np.where((t_labels-test_labels) == 0)[0]
                correctClassificationRate[i] = where.shape[0]/test_labels.shape[0] 
            else:
                continue
        maximum = np.max(correctClassificationRate)
        choosed_feature = np.where(correctClassificationRate == maximum)
        indices.remove(choosed_feature[0][0])
        CRR_BFS[j] = maximum
    indices.append(CRR_BFS[itr-1])
    return indices_BFS
```

```python
def SFS(train_data, test_data, test_labels, itr, indices):
    dim = train_data.shape[1]
    CCR_SFS = np.zeros(itr)
    for j in range(itr):
        correctClassificationRate = np.zeros((dim))
        for i in range(dim):
            if i not in indices:
                indx = indices.copy()
                indx.append(i)
                train = np.copy(train_data[:, indx])
                labels = np.copy(train_labels)
                test = np.copy(test_data[:, indx])
                gnb = MultinomialNB()
                t_labels = gnb.fit(train, labels).predict(test)
                where = np.where((t_labels-test_labels) == 0)[0]
                correctClassificationRate[i] = where.shape[0]/test_labels.shape[0]                      
            else:
                continue
        maximum = np.max(correctClassificationRate)
        choosed_feature = np.where(correctClassificationRate == maximum)
        indices.append(choosed_feature[0][0])
        CCR_SFS[j] =  maximum
    indices.append(CCR_SFS[itr-1])
    return indices
```

```python
indices=[]
itr = 30
CCR_SFFS = []
t0 = time()
while(True):
    if (len(indices) == itr):
               break
    elif (len(indices) == 0):
        ans = SFS (train_data, test_data, test_labels, 1, [])
        correct_classification_rate = ans[-1]
        indices = ans[:-1]
        flag = 1
        new_correct_classification_rate = correct_classification_rate
        CCR_SFFS.append(new_correct_classification_rate)
       
    elif (flag == 1):
        while(correct_classification_rate <= new_correct_classification_rate):
            correct_classification_rate = new_correct_classification_rate
            ans = SFS(train_data, test_data, test_labels, 1, indices)
            new_correct_classification_rate = ans[-1]
            CCR_SFFS.append(new_correct_classification_rate)
            indices = ans[:-1]
            flag = 0
            if (len(indices) == itr):
                break
        flag = 0
    elif(flag == 0):
        while(correct_classification_rate <= new_correct_classification_rate):
            correct_classification_rate = new_correct_classification_rate
            ans = BFS (train_data, test_data, test_labels, 1, indices)
            new_correct_classification_rate = ans[-1]
            CCR_SFFS.append(new_correct_classification_rate)
            indices = ans[:-1]
            flag = 0
            if (len(indices) == itr):
                break 
        flag = 1
print("Sequential floating forward selection choosing 30 features")
print("choosed features: ", indices)
print("done in %0.3fs" % (time() - t0))
               
with plt.style.context('seaborn-whitegrid'):
    
    plt.figure(figsize=(6, 4))
    plt.step(range(len(CCR_SFFS)), 1 - np.array(CCR_SFFS) , where='mid')
    plt.title("sequential floating forward selection choosing 30 features")
    plt.ylabel('probability of errore ')
    plt.xlabel('number of features')
    plt.tight_layout()
```
