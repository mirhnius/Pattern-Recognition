```python
import numpy as np
from time import time
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'png')
```

```python
dataset = fetch_lfw_people(min_faces_per_person=100)
n_samples, h, w = dataset.images.shape
data = dataset.data
labels = dataset.target
n_features = data.shape[1]
target_names = dataset.target_names
n_classes = target_names.shape[0]
```

```python
def plot_faces(images, n_row=2, n_col=5):
    plt.figure(figsize=(1.5 * n_col, 2.2 * n_row))
    plt.subplots_adjust(0.6, 0.5, 1.5, 1.5)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.xticks(())
        plt.yticks(())
    plt.tight_layout()
    plt.show()
```

```python
data_mean = np.mean(data,axis=0).reshape(1,n_features) # average face
#plot the average face and some samples from the dataset
plot_faces(np.concatenate((data_mean, data[5:15]),axis=0))
#plt.title('plot the average face and some samples from the dataset')
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.25, random_state=42)
```

```python
n_components = 50 
print("Extracting the top %d eigenfaces from %d faces"% (n_components, X_train.shape[0]))
t0 = time()
pca = PCA(n_components=n_components, svd_solver='randomized').fit(X_train)
print("done in %0.3fs" % (time() - t0))
```

```python
f,ax = plt.subplots()
plt.plot(pca.explained_variance_ratio_)
plt.title("ratio of explained variance")
ax.set_xlabel("principal component")
plt.show()
```

```python
Xmean = pca.mean_.reshape(1,n_features)
print("Projecting the input data on the eigenfaces orthonormal basis")
t0 = time()
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
print("done in %0.3fs" % (time() - t0))
```

```python
#pca components
fig, axes = plt.subplots(3, 8, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')
#original faces
plot_faces(X_test[5:15])
#reconstructed faces
X_test_rec = pca.inverse_transform(X_test_pca)
plot_faces(X_test_rec[5:15])
```
