```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
```

```python
test_data = np.loadtxt("fashion-mnist_test.csv", delimiter=',', unpack=True)
train_data = np.loadtxt("fashion-mnist_train.csv", delimiter=',', unpack=True)
```

```python
test_data = test_data.T
train_data = train_data.T
train_data = train_data[0:10000,:]
test_data = test_data[0:1000,:]
test_labels = test_data[:, 0] 
train_labels = train_data[:, 0] 
train_data = train_data[:, 1:]
test_data = test_data[:, 1:]

scaler = MinMaxScaler()
data = scaler.fit_transform(np.concatenate((test_data, train_data)))
test_data = data[0:1000, :]
train_data = data[1000:, :]
```

```python
CRR = []
for i in range(1,100,7):
    pca = PCA(n_components=i, svd_solver='randomized').fit(train_data)
    train_pca = pca.transform(train_data)
    test_pca = pca.transform(test_data)
    gnb = GaussianNB()
    t_labels = gnb.fit(train_pca, train_labels).predict(test_pca)
    CRR.append(accuracy_score(test_labels, t_labels))
```

```python
for i in range(100,755,17):
    pca = PCA(n_components=i, svd_solver='randomized').fit(train_data)
    train_pca = pca.transform(train_data)
    test_pca = pca.transform(test_data)
    gnb = GaussianNB()
    t_labels = gnb.fit(train_pca, train_labels).predict(test_pca)
    CRR.append(accuracy_score(test_labels, t_labels))
```

```python
indx = list(range(1,100,7))
indx2 = list(range(100,755,17))
indx = indx + indx2
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.plot(indx,CRR)  
    plt.figure(figsize=(6, 4))  
    plt.step(indx,CRR, where='mid')
    plt.ylabel('classifier performance')
    plt.xlabel('dimensionality')
    plt.tight_layout()
```
