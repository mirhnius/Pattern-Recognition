```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans
```

```python
dirname = os.path.dirname(os.path.abspath("__file__"))
filename = dirname + '/data'
os.chdir(filename) 
```

```python
df = pd.read_csv('ToyDataSet.csv') 
plebels = DBSCAN(eps=0.15, min_samples=5).fit_predict(df.values)
x = df.values[:,0]
y = df.values[:,1]
```

```python
cmap = plt.cm.rainbow
plt.figure(figsize=(14,7))
plt.scatter(x,y,c=plebels,cmap='viridis',s=50)
plt.title('DBSCAN', fontsize = 20)
plt.show()
```

```python
klebels = KMeans(n_clusters=10).fit_predict(df.values)
x = df.values[:,0]
y = df.values[:,1]
cmap = plt.cm.rainbow
```

```python
plt.figure(figsize=(14,7))
plt.scatter(x,y,c=klebels,cmap='viridis',s=50)
plt.title('k-means', fontsize = 20)
plt.show()
```
